### Now We Works On Every columns

# 1.Society


```python
data['society'].value_counts()
```

```python
data['society'].nunique()
```

```python
# Check For Null Values
data[data['society'].isnull()]
```

```python
# Fill Null value
data['society'] = data['society'].fillna('Not Specify')
```

```python
# Remove ratings from string
data['society'] = data['society'].apply(lambda x:re.sub(r'\d+(\.\d+)?\s*★', '', x).strip() )
```

```python
# convert society names into lower case
data['society'] = data['society'].apply(lambda x : x.lower())
```

# 2.Price

```python
data['price'].value_counts()
```

```python
data['price'].value_counts().shape
```

```python
data[data['price'] == 'Price on Request']
```

```python
# We drop rows where price = Price on Request	 because these rows doesnt help us to predict price
ind = data[data['price'] == 'Price on Request'].index
data.drop(ind,inplace=True)
```

```python
data[data['price'] == 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
data['price'] = data['price'].str.split(' ').apply(treat_price)
```

```python
data.head()
```

# 3.Price_per_square_fit

```python
data['price_per_sqft'] = data['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
