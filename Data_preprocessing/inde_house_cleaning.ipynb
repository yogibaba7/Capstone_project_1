```python
# Columns to drop -> property_name, link, property_id
data.drop(columns=['link','property_id'], inplace=True)
```

```python
data.head()
```

```python
# rename columns
data.rename(columns={'rate':'price_per_sqft'},inplace=True)
data.head()
```

# 1.Society

```python
data['society'].value_counts()
```

```python
data['society'].value_counts().shape
```

```python
data['society'] = data['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
data['society'].value_counts().shape
```

```python
data['society'] = data['society'].str.replace('nan','independent')
```

```python
data.head()
```

# 2.price

```python
data['price'].value_counts()
```

```python
data = data[data['price'] != 'Price on Request']
```

```python
data.head()
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
data['price'] = data['price'].str.split(' ').apply(treat_price)
```

```python
data.head()
```

# 3.price_per_sqft

```python

data['price_per_sqft'].value_counts()
```

```python
data['price_per_sqft'] = data['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
data.head()
```
